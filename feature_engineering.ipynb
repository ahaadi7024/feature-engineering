1. What is a parameter?

A parameter is a variable learned by a machine learning model during training. It determines how the model makes predictions.

Example: In Linear Regression, y=mx+c, the values m (slope) and c (intercept) are parameters.

2. What is correlation?

Correlation measures the strength and direction of the relationship between two variables.

Value ranges from -1 to +1
+1: Perfect positive correlation
0: No correlation
-1: Perfect negative correlation
3. What does negative correlation mean?

Negative correlation means that as one variable increases, the other decreases.

Example: Speed and travel time. Higher speed usually means lower travel time.

4. Define Machine Learning. What are the main components?

Machine Learning (ML) is a branch of AI that enables computers to learn patterns from data and make predictions without being explicitly programmed.

Main Components:

Data

5. How does loss value help determine whether a model is good?

Loss measures the difference between actual and predicted values.

Lower loss → Better predictions
Higher loss → Poor predictions

A good model generally has low training and testing loss.

6. What are continuous and categorical variables?
Continuous Variables

Numeric values that can take any value within a range.

Examples: Height, Weight, Temperature

Categorical Variables

Variables representing categories.

Examples: Gender, Color, City

7. How do we handle categorical variables?

Common techniques:

Label Encoding
Assign numbers to categories.
Example: Red=0, Blue=1, Green=2
One-Hot Encoding
Create binary columns.
Example: Color_Red, Color_Blue
Ordinal Encoding
Used when categories have an order.
Example: Low, Medium, High

15. What is causation? Difference between correlation and causation?
Correlation

Two variables move together.

Causation

One variable directly causes changes in another.

Example:
Ice cream sales and drowning incidents increase in summer.

Correlation exists.
Ice cream does not cause drowning.
Hot weather causes both.
16. What is an Optimizer?

An optimizer updates model parameters to minimize the loss function.

Types of Optimizers
Gradient Descent

Updates parameters using the entire dataset.

θ = θ - α∇J(θ)
Stochastic Gradient Descent (SGD)

Updates after each training example.

Mini-Batch Gradient Descent

Updates after a small batch of samples.

Adam

Combines momentum and adaptive learning rates.
Most widely used optimizer.

optimizer='adam'
RMSProp

Uses adaptive learning rates.

Adagrad

Adjusts learning rate based on feature frequency.

20. What is feature scaling? How does it help?

Feature scaling standardizes feature values so that all features contribute equally.

Benefits:

Faster convergence
Improved accuracy
Better performance for distance-based algorithms

22. Explain Data Encoding

Data encoding converts categorical data into numerical form so machine learning algorithms can process it.

Label Encoding
from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()
df['Gender'] = le.fit_transform(df['Gender'])
One-Hot Encoding
import pandas as pd

encoded_df = pd.get_dummies(df, columns=['Gender'])

Purpose: Convert categorical features into machine-readable numerical values
